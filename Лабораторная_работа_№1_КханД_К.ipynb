Сначала откроем выбранный датасет. Загрузим таблицу в переменную data и выведем ее. Укажем, что разделителем по столбцам в csv файле является запятая

Выведем первый и третий столбцы выборки сначала вместе, потом по отдельности.
Функция iloc позволяет выбирать объекты таблицы по индексам. Первый аргумент - индекс или срез строк, второй - индекс или срез столбцов

```python
data.iloc[:, 1:4:2]
```

По отдельности

```python
data.iloc[:,1]
data.iloc[:,3]
```

Выведем первые 10 строк выборки, используя тот же метод iloc

```python
data.iloc[:10, :]
```

Выведем 23 случайных строки из выборки, используя метод sample

Выведем статистические данные из числовых столбцов. Используя метод describe()

```python
data.describe()

```

выведем разницу в статистическом поле в данных для женщин и мужчин, так как на основе этих данных можно будет сделать некоторые выводы

```python

men = data[data['Gender'] == 'Male']
women = data[data['Gender'] == 'Female']
men.describe()
women.describe()
```

Сделаем базовую группировку по полу

```python
grouped_by_gender = data.groupby('Gender').agg({
    'Height': ['mean', 'std', 'min', 'max'],
    'Weight' : ['mean', 'std', 'min', 'max'],
    'Age' : ['mean', 'std', 'min', 'max'],
    'BMI' : ['mean', 'std', 'min', 'max'],
})
grouped_by_gender
```

Посмотрим количество пустых значений (Nan) в столбцах

Описание атрибутов выборки на русском:
Антропометрические данные: возраст,рост, вес
Медицинские показатели: индекс массы тела, категория индекса массы тела, процент жира в организме
Физиологические показатели: суточное потребление калорий, сердечный ритм, метаболический возраст
Демографические и поведенческие показатели: пол, уровень активности, рекомендуемая диета

Графически проверим датасет на наличие выбросов. Визуализировать будем с помощью boxplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Настройка стиля
plt.style.use('default')
sns.set_palette("husl")

# Boxplot для всех числовых столбцов
numeric_columns = data.select_dtypes(include=[np.number]).columns

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()

for i, col in enumerate(numeric_columns):
    if i < len(axes):
        data.boxplot(column=col, ax=axes[i])
        axes[i].set_title(f'Выбросы в {col}', fontweight='bold')
        axes[i].tick_params(axis='x', rotation=45)

plt.suptitle('ВИЗУАЛИЗАЦИЯ ВЫБРОСОВ В ЧИСЛОВЫХ СТОЛБЦАХ', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
